```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture

from sklearn.datasets import make_moons
```

```python
X  , y= make_circles(n_samples = 500 , noise=0.05)
```

```python
def plot(X , y_pred_KM,y_pred_sp,
         y_pred_DB,y_pred_Agg,y_pred_G ):
    plt.figure(figsize = (20,5))

    plt.subplot(161)
    plt.title('Data')
    plt.scatter(X[: ,0] ,X[: ,1] , c='gray')



    plt.subplot(162)
    plt.title('Kmean')
    plt.scatter(X[y_pred_KM == 1,0] , X[y_pred_KM == 1,1]);
    plt.scatter(X[y_pred_KM == 0,0] , X[y_pred_KM == 0,1]);



    plt.subplot(163)
    plt.title('Spectural')
    plt.scatter(X[y_pred_sp == 1,0] , X[y_pred_sp== 1,1]);
    plt.scatter(X[y_pred_sp == 0,0] , X[y_pred_sp == 0,1]);



    plt.subplot(164)
    plt.title('DBscan')
    plt.scatter(X[y_pred_DB == 1,0] , X[y_pred_DB== 1,1]);
    plt.scatter(X[y_pred_DB == 0,0] , X[y_pred_DB == 0,1]);



    plt.subplot(165)
    plt.title('Agglo')
    plt.scatter(X[y_pred_Agg == 1,0] , X[y_pred_Agg== 1,1]);
    plt.scatter(X[y_pred_Agg == 0,0] , X[y_pred_Agg == 0,1]);

    plt.subplot(166)
    plt.title('Gaussian')
    plt.scatter(X[y_pred_G == 1,0] , X[y_pred_G== 1,1]);
    plt.scatter(X[y_pred_G == 0,0] , X[y_pred_G == 0,1]);
```

```python
plot(X , y_pred_KM,y_pred_sp,
         y_pred_DB,y_pred_Agg,y_pred_G )
```

```python
X  , y= make_circles(n_samples = 500 , noise=0.05)
```

```python
X[y==0,:] = 2*X[y==0 ,:]
```

```python
X.shape
```

```python
plt.scatter(X[y==1,0] , X[y==1,1]);
plt.scatter(X[y==0,0] , X[y==0,1])
```

```python
y_pred_KM= KMeans(n_clusters = 2).fit_predict(X)
y_pred_sp=SpectralClustering(n_clusters = 2).fit_predict(X)
y_pred_DB=DBSCAN(eps = 0.3, min_samples=3).fit_predict(X)
y_pred_Agg=AgglomerativeClustering(n_clusters=2).fit_predict(X)
y_pred_G = GaussianMixture(n_components= 2).fit_predict(X)
```

```python
plt.figure(figsize = (20,5))

plt.subplot(161)
plt.title('Data')
plt.scatter(X[: ,0] ,X[: ,1] , c='gray')



plt.subplot(162)
plt.title('Kmean')
plt.scatter(X[y_pred_KM == 1,0] , X[y_pred_KM == 1,1]);
plt.scatter(X[y_pred_KM == 0,0] , X[y_pred_KM == 0,1]);



plt.subplot(163)
plt.title('Spectural')
plt.scatter(X[y_pred_sp == 1,0] , X[y_pred_sp== 1,1]);
plt.scatter(X[y_pred_sp == 0,0] , X[y_pred_sp == 0,1]);



plt.subplot(164)
plt.title('DBscan')
plt.scatter(X[y_pred_DB == 1,0] , X[y_pred_DB== 1,1]);
plt.scatter(X[y_pred_DB == 0,0] , X[y_pred_DB == 0,1]);



plt.subplot(165)
plt.title('Agglo')
plt.scatter(X[y_pred_Agg == 1,0] , X[y_pred_Agg== 1,1]);
plt.scatter(X[y_pred_Agg == 0,0] , X[y_pred_Agg == 0,1]);

plt.subplot(166)
plt.title('Gaussian')
plt.scatter(X[y_pred_G == 1,0] , X[y_pred_G== 1,1]);
plt.scatter(X[y_pred_G == 0,0] , X[y_pred_G == 0,1]);
```
